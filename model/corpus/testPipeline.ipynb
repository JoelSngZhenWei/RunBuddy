```python
import os, re, json, uuid, hashlib, urllib.parse
from datetime import datetime, timezone
from typing import Dict, Any, List
import requests

from docling.document_converter import DocumentConverter
```

```python
TEST_URL = "https://en.wikipedia.org/wiki/Marathon"  # change if you want
TEST_LICENSE = "CC-BY-SA 4.0"
TEST_SOURCE  = "wikipedia"
TEST_TAGS    = ["marathon","endurance","training"]

print(TEST_URL)
```

```python
converter = DocumentConverter()
result = converter.convert(TEST_URL)
```

```python
md = result.document.export_to_markdown()
md_clean = re.sub(r"\s*\[[^\]]+\]", "", md)
md_clean = re.split(r"\n#+\s*(References|See also)\b", md_clean)[0].strip()
```
